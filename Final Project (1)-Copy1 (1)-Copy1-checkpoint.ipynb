```python
def crossvalidation(X,k,length,max_depth): #k- no of fold
    lenoffold=int(length/k)
    folddataset=[]
    acc=[]
    for i in range(k):
        l=[]
        for i in range(lenoffold):
            a=random.randrange(len(X))
            l.append(X[a][:])
        #print(len(l))
        folddataset.append(l)
    #print('f',len(folddataset))
    for j in range(k):
        #print('out',j)
        acc.append(traintestsplit(folddataset,j,k,max_depth))
    print('Accuracy', acc)
    return acc
```

```python
def calculate_entropy(labels):
    # Calculate entropy of a set of labels
    unique_labels, counts = np.unique(labels, return_counts=True)
    probabilities = counts / len(labels)
    entropy = -np.sum(probabilities * np.log2(probabilities))
    return entropy

def split_data(data, labels, feature_index, threshold):
    # Split data and labels based on a threshold for a specific feature
    left_data, left_labels = data[data.iloc[:, feature_index] <= threshold], labels[data.iloc[:, feature_index] <= threshold]
    right_data, right_labels = data[data.iloc[:, feature_index] > threshold], labels[data.iloc[:, feature_index] > threshold]
    return left_data, left_labels, right_data, right_labels

def find_best_split(data, labels):
    # Find the best feature and threshold to split the data
    best_entropy = float('inf')
    best_feature_index = None
    best_threshold = None
    
    for feature_index in range(data.shape[1]):
        unique_values = np.unique(data.iloc[:, feature_index])
        for threshold in unique_values:
            left_data, left_labels, right_data, right_labels = split_data(data, labels, feature_index, threshold)
            if len(left_labels) == 0 or len(right_labels) == 0:
                continue
            
            left_entropy = calculate_entropy(left_labels)
            right_entropy = calculate_entropy(right_labels)
            combined_entropy = (len(left_labels) / len(labels)) * left_entropy + (len(right_labels) / len(labels)) * right_entropy
            
            if combined_entropy < best_entropy:
                best_entropy = combined_entropy
                best_feature_index = feature_index
                best_threshold = threshold
    
    return best_feature_index, best_threshold

def build_tree(data, labels, depth=0, max_depth=None):
    # Recursive function to build the decision tree
    if max_depth is not None and depth >= max_depth:
        return Node(data, labels)
    
    if len(np.unique(labels)) == 1:
        return Node(data, labels)
    
    best_feature_index, best_threshold = find_best_split(data, labels)
    if best_feature_index is None:
        return Node(data, labels)
    
    left_data, left_labels, right_data, right_labels = split_data(data, labels, best_feature_index, best_threshold)
    
    left_subtree = build_tree(left_data, left_labels, depth + 1, max_depth)
    right_subtree = build_tree(right_data, right_labels, depth + 1, max_depth)
    
    node = Node(data, labels)
    node.feature_index = best_feature_index
    node.threshold = best_threshold
    node.left = left_subtree
    node.right = right_subtree
    
    return node
```

```python
def predict(tree, example):
    # Traverse the tree to predict the label for an example
    while tree.left or tree.right:
        if example[tree.feature_index] <= tree.threshold:
            tree = tree.left
        else:
            tree = tree.right
    return tree.value
```

```python
def accuracyscore(ypred,yactual):
    ytest=np.array(yactual)
    ypre=np.array(ypred)
    return np.sum(ytest == ypre) / len(ytest)
```

```python
class Node:
    def __init__(self, data, labels):
        self.data = data
        self.labels = labels
        self.feature_index = None
        self.threshold = None
        self.left = None
        self.right = None
        self.value = np.mean(labels) 
```
